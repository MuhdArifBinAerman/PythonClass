```python
# Square

T.initializeTurtle() # Create a new turtle and discard the recent one
T.forward(100)
T.right(90)
T.forward(100)
T.right(90)
T.forward(100)
T.right(90)
T.forward(100)
T.right(90)
```

---
Noticed the repeated code from the last cell ‚ùì

In programming, repeated codes are encourage to be eliminated as it makes the programme more difficult to understand.

To improve the code above, we can simply use `for` loop.

```python
# Square
T.initializeTurtle()

for i in range(4):
  T.forward(100)
  T.right(90)
```

Recall:

Remember the " **Loops** " session just now ‚ùì

When we know how many times we need our code to be executed, we use `for` loop.

In this case, we need to execute `forward()` and `right()` 4 times respectively, thus we set a range for the programme to execute them 4 times.

## Circle in ColabTurtle

ColabTurtle doesn't provide us with circle function. In order to create a circle, we will define our own function with the logic to draw a circle.

Below is the function for creating a circle.

<h4> Credits : <a href:"https://stackoverflow.com/questions/59616766/google-colaboratory-attributeerror-module-colabturtle-turtle-has-no-attribut"> Stack Overflow </a> </h4>

```python
from math import pi
def tcircle(radius):
    T.speed(12)
    #function could be summarized into:
    #regular_polygon(int((2 * pi * radius)/9)),9)

    #explained step by step:
    """draws a regular polygon of n sides
    that is supposed to appear like a circle.
    n is set to 9 for fast drawing time.
    it calculates rounded side length from n and radius"""
    #circumference (c)= 2*pi*radius
    c = 2 * pi * radius


    #n = amount of lines or corners, it defines the accuracy of the circle
    n = 30 # lower number to decrease drawing time (can be any float or int)

    #circumference (c) = ca.  l * n
    #l = length of individual lines 
    l = c / n

    regular_polygon(int(l),n)


def regular_polygon(l, n):
    """draws a regular polygon of n amount sides of length l
    that is supposed to appear like a circle.
    function by cdlane from a stackoverflow post"""
    for _ in range(n):
        T.forward(l)
        T.left(360 / n)
```

Simple circle

```python
T.initializeTurtle()
tcircle(50)
```

Draw several circles with `for` loop. ‚≠ï

```python
T.initializeTurtle()
for i in range(12):
  if (i%2 == 0):
    T.color('red')
  else:
    T.color('orange')
  tcircle(50)
  T.rt(30)
```
