<h1>The topic taught below is called Nested Functions or Inner Functions. It provides encapsulation and makes it harder for people to change the function.</h1>
<h3>It might be difficult for beginners to learn. Please do not hesistate to seek clarification if needed.</h3>

```python
# Nested function is referred to as "functions containing another function"

# Example
def outer_func(param1):
  print(param1)
  def inner_func():
    pass

outer_func(3) #3 is outputted
```

```python
# Notice if the print function is placed inside inner_func(), nothing will be displayed

# Example
def outer_func(param1):
  def inner_func():
    print(param1)

outer_func(3) 
```

```python
# This is because the print statement is located inside inner_func, so we need to call the inner_func inside the outer_func to see the output
# Example
def outer_func(param1):
  def inner_func():
    print(param1)
  inner_func()

outer_func(3) 
```

```python
# Returns an error if we call inner_func(). The program is successfully protected from external modifications.
inner_func()
```
