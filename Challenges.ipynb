````python
#@title Creating an Excel file
from openpyxl import Workbook
workbook = Workbook()

workbook.save(filename="hello_world.xlsx") #Important, used to save the excel file
````

<h1>Write Excel File</h1>

````python
# Writing an Excel file
import openpyxl

# Creates workbook which is then saved as Excel
wb = openpyxl.Workbook()
sheet = wb.active

# Cells have row and column as their coordinates
# Note: The first row or column integer is 1, not 0
# This cell will be A1
c1 = sheet.cell(row = 1, column = 1) 

# Assigning the string "Hello" to the cell
c1.value = "Hello" 

# This cell will be B1
c2 = sheet.cell(row = 1, column = 2)
c2.value = "World!"

# Instead of coordinates, you can also refer to a cell by name
c3 = sheet["A2"]
c3.value = "Welcome"
c4 = sheet["B2"]
c4.value = "Everyone"

# Remember to save everytime
wb.save("hello_world.xlsx")
````

<h1>Read Excel File</h1>


````python
import openpyxl

# Opens Excel file
excelFile = openpyxl.load_workbook("hello_world.xlsx")
# Opens the first available sheet
excelSheet = excelFile.active
# Takes value from cell
A1 = excelSheet.cell(row = 1, column = 1) 

print(A1.value) 

Desired = excelSheet.cell(row = 4, column = 1)

print(Desired.value)
````

<h1>Append Excel file</h1>

````python
import openpyxl

# Opens existing Excel file instead of creating a new blank workbook
excelFile = openpyxl.load_workbook("hello_world.xlsx")
sheet = excelFile.active

# Assigning a cell a value
A3 = sheet["A3"]
A3.value = "New Data"

excelFile.save("hello_world.xlsx")
````

````python
import openpyxl

excelFile = openpyxl.load_workbook("hello_world.xlsx")
sheet = excelFile.active

# Using Append() adds data to the end of the sheet
data = (
    (1, 2, 3),
    (4, 5, 6)
)

for row in data:
  sheet.append(row)

excelFile.save("hello_world.xlsx")
````

````python
import openpyxl

excelFile = openpyxl.load_workbook("hello_world.xlsx")
sheet = excelFile.active

# Select cells from A1 to C5
A4_C5 = sheet["A1" : "C5"]

# Column A, B and C
for A, B, C in A4_C5:
    print(A.value, B.value, C.value)
````

<h1>Create PDF</h1>

````python
from fpdf import FPDF
 
# create a object named pdf 
pdf = FPDF()
 
# add a page (need this to be added or else pdf cannot be created)
pdf.add_page()
 
# save the pdf with name .pdf
pdf.output("testing.pdf")  
````

<h1>Write PDF</h1>

````python
# create a object named pdf 
pdf = FPDF()

# set style and size of font
# that you want in the pdf
pdf.set_font("Arial", size = 15)
 
# add a page (need this to be added or else cannot create cells)
pdf.add_page()

# create cell
pdf.cell(10, 10, txt = "Hello", #x starting pos is 10 and y starting pos is 10
         ln = 1, align = 'C') #ln means line and align is how you want the text to placed in the pdf
 
# add cell
pdf.cell(20, 20, txt = "How are you?", #x starting pos is 20 and y starting pos is 20
         ln = 2, align = 'C') #ln means line and align is how you want the text to placed in the pdf
 
# save the pdf with name .pdf
pdf.output("testing1.pdf") 
````

<h1>Read PDF</h1>

````python
# install PyPDF2 library to create PDF document
!pip install PyPDF2
````

````python
# importing required libraries
from PyPDF2 import PdfReader

reader = PdfReader("testing1.pdf")
text = ""
for page in reader.pages:
    text += page.extract_text() + "\n"
    
print(text)
````

<h1><b>Dictionary Section</b></h1>
<h3>In this section, we will cover the usage of dictionaries. This might be a difficult topic so do not hesitate to seek for clarification.</h3>

````python
# You can choose to create a dictionary by initializing or initializing and assigning key-value pairs immediately

# Initialize a dictionary only
first_dict = {}

# Initializing and assigning
person = {"John": 31, "Jane": 29, "Chloe": 23, "Melvin": 19}
````

````python
# To assign an initialized dictionary, we use the following code

# The LHS specifies the key of the dictionary while the RHS specifies the value of the dictionary
first_dict["key"] = "value"

first_dict
````

````python
# We can add any key-value pairs in any dictionary
first_dict["age"] = 3
first_dict["name"] = "Guang Yu"
first_dict # Notice they are appended to the end of the dictionary
````

````python
# To update the value for a key
first_dict["age"] = 10
first_dict # Notice the value for age has changed
````

````python
# To extract each key and value in the dictionary
for key, value in person.items(): #items is a function that is used to the get each key and value pair in the dictionary
  print("Name:", key, "Age:" , value)
````

<h2>Tuple

Tuples is a collection of data.

It is written with round brackets.

```
new_tuple = ("hello","welcome","have fun")

```

<br />

Tuples items:

- Ordered
  - The order is fixed

- Unchangeable
  - Cannot change, add or remove

- Allows duplicates
  - Allows item with same value

<br />

<h2>One-item Tuple

To create a tuple with only 1 item in it, a " , " is required for Python to recognize it as a tuple.

````python
tuple1 = ("hey")      # this is a string, not a tuple
print(type(tuple1))

tuple1 = ("hey",)     # this is a tuple
print(type(tuple1))
````

<br />
