### While Loops

**While loop** allow you to excute the instructions in it until the condition is `false`.


How to construct a while loop üîΩ:
```
while (condition):
  (your code)
```





````python
# Uncomment the following lines to see the output

# a = 1

# while (a <= 10):
#   print(a)
#   a = a + 1
````

From the code above,

* We set the variable " a " with the value 1.
* We use a `while` loop to check whether " a " is smaller than or equal to 10.
* If yes, we `print` the value of " a ".

Please take note of the `a = a + 1`. We don't wish the loop to continue forever ( or in python, we call it " **infinite loop** " ), so we make some changes on the variable a so it won't forever holds the value of 1. In our case, we increase the value of " a " by 1 every iteration. When " a " reaches the value 11, the condition become `false` and `while` loop stops.

## Differences (`for` vs `while`)

We usually use `for` loop when we know how many times we wish to execute the code. On the other hand, when we not sure how many times we want to execute it , we use `while` loop. " **While** " loop only stops when the condition is `false`.

## Continue


Keyword `continue` will skip the current iteration and move to the next iteration.

`if` statement here help us to define our condition.

````python
# Uncomment the following lines to see the output

a = 1

while (a <= 10):
  a = a + 1 
  if (a == 4):      # When a holds the value of 4, we escape from the loop
    continue        # 4 will not be printed
  print(a)
    
````

From the output above, why the value starts from 2 ‚ùì

Python execute the code line by line.

In the first iteration, " a " holds the value of 1 when going into the `while` loop.

However, we increment it by 1 before printing it.

By the time it reaches the `print` function, " a " now holds the value of 2 instead.

---

Another example with `for` loop.

````python
# Uncomment the following lines to see the output

for i in range(10):
  if (i == 4):      # When i holds the value of 4, skip the current iteration
    continue        # 4 will not be printed
  print(i)
 
````
