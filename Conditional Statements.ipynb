## Keyword `elif`

Keyword `elif` is usually placed after `if`.

Meaning that check this ( `elif` ) if the statement above ( `if` ) is wrong.

````python
# Please take note of the indentation
# Uncomment the following lines to see the output

a = 10
b = 10

if (a > b):
  print("a is greater than b")

elif (a == b):
  print("a is equal to b")
````

Condition for `if` will be check first before going to `elif`.

````python
a = 10
b = 10

if (a == b):
  print("a is equal than b")

elif (a > b):
  print("a is greater to b")
````

## Keyword `else`

`else` allows you to give action when none of the conditions is met

````python
# Uncomment the following lines to see the output

a = 10
b = 10

if (a > b):
  print(" a is greater than b ")
else:
  print(" no condition is met ")
````

You can add an `elif` in between.


````python
# Uncomment the following lines to see the output

a = 10
b = 10

if (a > b):
  print("a is greater than b")
elif (a < b):
  print("a is smaller than b")
else:
  print("no condition is met")
````

And not only 1 `elif` if you wish to use more.

````python
# Uncomment the following lines to see the output

a = 10
b = 10

if (a > b):
  print("a is greater than b")
elif (a < b):
  print("a is smaller than b")
elif (a == 20):
  print("a is 20")
elif (b == 30):
  print("b is 30")
else:
  print("no condition is met")
````

As mentioned earlier, `if`, `elif` and `else` statement are checked accoridingly from `if` to `else`. 

Once a condition is met, the following conditions will not be checked.

````python
# Uncomment the following lines to see the output

a = 10
b = 10

if (a == 10):
  print("a is equal to 10")
elif (b == 10):
  print("b is equal to 10")
elif (a == b):
  print("a is equal to b")
else:
  print("no condition is met")
````
